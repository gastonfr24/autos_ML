```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('car.csv', header= None)
```

```python
data.columns= ['Price','Maintenence Cost','Number of Dosrs',
               'Capacity','Size of Luggage Boot','Safety','Decision']
```

```python
data.head()
```

```python
decision = data['Decision'].value_counts()
```

```python
decision.plot(kind = 'bar') 
```

```python
data['Price'].unique()
```

```python
data['Price'].replace(('vhigh', 'high', 'med', 'low'),(4,3,2,1), inplace= True)
```

```python
data['Price'].unique()
```

```python
price = data['Price'].value_counts()
```

```python
colors= ['#DDEE01','#FF0101','#FE10D1','#0000FF']
price.plot(kind = 'bar', color = colors)
plt.xlabel('Precios')
plt.ylabel('Autos')
plt.title('Precios de Autitos')
```

```python
labels= ['low','med','high']
```

```python
size= [576, 576, 576]
colors= ['yellow','green','red']
explode= [0, 0, 0]
```

```python
plt.pie(size,labels = labels, colors= colors, explode = explode, shadow= True, autopct = '%.2f%%')
plt.title('Niveles de seguridad', fontsize= 10)
plt.axis('off')
plt.legend(loc = 'best')
plt.show()
```
