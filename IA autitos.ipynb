```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= pd.read_csv('car.csv', header= None)
```

```python
data.columns= ['Price','Maintenence_Cost','Number_of_Dosrs',
               'Capacity','Size_of_Luggage_Boot','Safety','Decision']
```

```python
data.sample(3)
```

```python
data['Decision'].value_counts()
```

```python
data.Price.replace(('vhigh','high','med','low'),(4,3,2,1), inplace= True)
data.Maintenence_Cost.replace(('vhigh','high','med','low'),(4,3,2,1), inplace= True)
data.Number_of_Dosrs.replace(('2','3','4','5more'),(1,2,3,4), inplace= True)
data.Capacity.replace(('2','4','more'),(1,2,3), inplace= True)
data.Size_of_Luggage_Boot.replace(('big','med','small'),(3,2,1), inplace= True)
data.Safety.replace(('high','med','low'),(3,2,1), inplace= True)
data.Decision.replace(('unacc','acc','good','vgood'),(1,2,3,4), inplace= True)
```

```python
data.head()
```

```python
dataset= data.values
X= dataset[:, 0:6]
Y= np.asarray(dataset[:,6],dtype= 'S6')
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size= 0.2, random_state=0)
```

```python
tr= tree.DecisionTreeClassifier(max_depth=10)
```

```python
tr.fit(X_Train,Y_Train)
```

```python
y_pred = tr.predict(X_Test)
```

```python
print(y_pred)
```

```python
score = tr.score(X_Test,Y_Test)
```

```python
print('Presición: %0.3f' % (score))
```
